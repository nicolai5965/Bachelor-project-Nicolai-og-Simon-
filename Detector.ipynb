```python
def cyl(x,y,z):
    r = np.sqrt(x**2 + y**2)
    phi = np.arctan2(y,x)
    z = z
    return r , phi , z

rd, pd, zd = cyl(det.cx, det.cy, det.cz)
det['r'] = rd
det['phi'] = pd
lr, lp, lz = cyl(layers.cx, layers.cy, layers.cz)
layers['r'] = lr
layers['phi'] = lp
```
